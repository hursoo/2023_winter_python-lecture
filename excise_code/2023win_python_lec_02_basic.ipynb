## 5. 변수, 파생변수

### 1) 변수명 바꾸기: 
- 복사본 df를 이용해 manufacturer을 company로 변경하라

```python
mpg_new = mpg.copy()
mpg_new = mpg_new.rename(columns = {'manufacturer' : 'company'})
mpg_new
```

### 2) 파생변수 만들기
- mpg에서 cty, hwy 두 변수를 더해 2로 나눠 도로 유형을 통합한 연비 변수를 만들어 total 변수에 저장

```python
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg.head(3)
```

### 3) 조건문 사용
- total이 20 이상이면 pass, 그렇지 않으면 fail을 부여 -> test 변수에 저장

```python
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg.head(3)
```

## 6. df에서 빈도 확인하기

### 1) 빈도표 만들기
- 빈도표로 합격 판정 자동차 수 살펴서 변수 count_test에 저장

```python
count_test = mpg['test'].value_counts()
count_test
```

### 2) 빈도 막대 그래프 만들기
- 합격 판정 자동차 빈도표로 빈도 막대 그래프를 만들어라

```python
count_test.plot.bar()
```
